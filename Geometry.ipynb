## 1. Max Points on a Line (LC149)
```
Given n points on a 2D plane, find the maximum number of points that lie on the same straight line.
//  HashMap<Double, Integer> hm = new HashMap<Double, Integer>();
// hm.length        if(hm.containsKey(x)){.....}         hm.get()        hm.put(k, v)       Math.max
class Solution {
    public int maxPoints(Point[] points) {
        if(points.length <= 0) return 0;
        if(points.length <= 2) return points.length;
        int result = 0;
        for(int i = 0; i < points.length; i++){
            HashMap<Double, Integer> hm = new HashMap<Double, Integer>();
            int samex = 1;                   // infinite slope
            int samep = 0;                   // same point
            for(int j = 0; j < points.length; j++){
                if(j != i){
                    if((points[j].x == points[i].x) && (points[j].y == points[i].y)){
                        samep++;
                    }
                    if(points[j].x == points[i].x){
                        samex++;
                        continue;
                    }
                    double k = (double)(points[j].y - points[i].y) / (double)(points[j].x - points[i].x);      // Casting
                    if(hm.containsKey(k)){
                        hm.put(k,hm.get(k) + 1);
                    }else{
                        hm.put(k, 2);
                    }
                    result = Math.max(result, hm.get(k) + samep);
                }
            }
            result = Math.max(result, samex);
        }
        return result;
    }
}
```
## 2. Cross Product
#### The most important operation in 2D geometry
#### Define ccw(A, B, C) = (B − A) × (C − A) = (bx − ax)(cy − ay) − (by − ay)(cx − ax) <br> if ccw(A,B,C) > 0, ABC is a left turn, else right turn
**Example** Segment-Segment Intersection Test <br> Given two segment AB and CD, do they segments? <br>
The intersection exsits iff ccw(ABC) ccw(ABD) < 0  AND  ccw(CDA) ccw(CDB) < 0

## 3. Convex Hull Problem
Given n points on the plane, find the smallest convex polygon that contains all the given points
#### 3.1 Simple Algorithm
AB is an edge of the convex hull iff ccw(ABC) have the same sign for all other points C
Complexity  O(N^3)
#### 3.2 Faster Algorithm
1. The left most point must be in the region
2. Sort the points in increasing order of y/x
3. Incrementally construct the convex hull using a stack
```
Set the leftmost point as (0, 0), and sort the rest of the points in increasing order of y/x 
◮ Initialize stack S
◮ For i = 1, . . . , n 
    – Let A be the second topmost element of S, B be the topmostelement of S, and C be the ith point
    – If ccw(A, B, C) < 0, pop S and go back
    – Push C to S
◮ Points in S form the convex hull
```

## 5. Intersecting Half-planes
> Representing a half-plane: ax + by + c ≤ 0
> Given n half-planes, find the intersection of them
#### 5.1 Simple Algorithm O(n^3)

```
For each half-plane aix + biy + ci ≤ 0, define a straight line ei: aix + biy + ci = 0
    ◮ For each pair of ei and ej :
        – Compute their intersection p = (px, py)
        – Check if akpx + bkpy + ck ≤ 0 for all half-planes
    ◮ If so, store p in some array P
    ◮ Otherwise, discard p
    ◮ Find the convex hull of the points in P
```
